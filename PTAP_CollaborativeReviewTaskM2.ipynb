```
#@title 3. Calculate Statistics (mean, standard deviation, skewness, excess kurtosis)
import scipy.stats as stats

tesla_mean = tesla['Returns'].mean()
tesla_std = tesla['Returns'].std()
tesla_skew = stats.skew(tesla['Returns'].dropna())
tesla_kurtosis = stats.kurtosis(tesla['Returns'].dropna())

exxon_mean = exxon['Returns'].mean()
exxon_std = exxon['Returns'].std()
exxon_skew = stats.skew(exxon['Returns'].dropna())
exxon_kurtosis = stats.kurtosis(exxon['Returns'].dropna())

print('Tesla')
print(f'The mean of Tesla Returns is: {tesla_mean}')
print(f'The standard deviation of Tesla Returns is: {tesla_std}')
print(f'The skewness of Tesla Returns is: {tesla_skew}')
print(f'The kurtosis of Tesla Returns is: {tesla_kurtosis}')

print('\nExxon Mobil')
print(f'The mean of Exxon Mobil Returns is: {exxon_mean}')
print(f'The standard deviation of Exxon Mobile Returns is: {exxon_std}')
print(f'The skewness of Exxon Mobil Returns is: {exxon_skew}')
print(f'The kurtosis of Exxon Mobil Returns is: {exxon_kurtosis}')
```

```
#@title 4. Calculate Covariance
covariance_tesla_exxon = np.cov(tesla['Returns'].dropna(),
                                exxon['Returns'].dropna())[0][1]

print(f"The Covariance between Tesla and Exxon Mobile is: {covariance_tesla_exxon}")
```

### States 
for the portfolio with $80\%$ and the portfolio with $20\%$

```
#@title portfolio_80 Categorizing each day UU, UD, DU, DD
portfolio_80 = tesla_80.copy()
portfolio_80['Ex_returns'] = exxon_80['Returns']
portfolio_80.rename(columns={'Returns': 'Ts_returns'}, inplace=True)
portfolio_80 = portfolio_80[['Ts_returns', 'Ex_returns']]
portfolio_80['Ts_UD'] =  np.where(portfolio_80['Ts_returns'] > 0, 'U', 'D')

portfolio_80['Ex_UD'] =  np.where(portfolio_80['Ex_returns'] > 0, 'U', 'D')
portfolio_80['Joint_movt'] =  portfolio_80['Ts_UD']+portfolio_80['Ex_UD']

portfolio_80['Previous_movt'] = portfolio_80['Joint_movt'].shift(1)

portfolio_80.tail()
```

```
#@title portfolio_20 Categorizing each day UU, UD, DU, DD
portfolio_20 = tesla_20.copy()
portfolio_20['Ex_returns'] = exxon_20['Returns']
portfolio_20.rename(columns={'Returns': 'Ts_returns'}, inplace=True)
portfolio_20 = portfolio_20[['Ts_returns', 'Ex_returns']]
portfolio_20['Ts_UD'] =  np.where(portfolio_20['Ts_returns'] > 0, 'U', 'D')

portfolio_20['Ex_UD'] =  np.where(portfolio_20['Ex_returns'] > 0, 'U', 'D')
portfolio_20['Joint_movt'] =  portfolio_20['Ts_UD']+portfolio_20['Ex_UD']

portfolio_20['Previous_movt'] = portfolio_20['Joint_movt'].shift(1)

portfolio_20.tail()
```

### Directions
for the portfolio with $80\%$ and the portfolio with $20\%$

```
#@title Direction for Portfolio 80
portfolio_20['Direction'] =  portfolio_20['Previous_movt'] + '_' + portfolio_20['Joint_movt']
portfolio_20[['Joint_movt',	'Previous_movt', 'Direction']]
portfolio_80['Direction'] =  portfolio_80['Previous_movt'] + '_' + portfolio_80['Joint_movt']
portfolio_80[['Joint_movt',	'Previous_movt', 'Direction']]
```

```
#@title Direction for Portfolio 20
portfolio_20['Direction'] =  portfolio_20['Previous_movt'] + '_' + portfolio_20['Joint_movt']
portfolio_20[['Joint_movt',	'Previous_movt', 'Direction']]
```

### Transition Matrix

```
#@title Creating a transition Matrix
def movement(status, p_status):
    if (status == "UU" or status == "DD"):
        if (p_status == "UU" or p_status == "DD"):
            result = "mt_mt"
        else:
            result = "mt_ma"
    else:
        if (p_status == "UU" or p_status == "DD"):
            result = "ma_mt"
        else:
            result = "ma_ma"
    return result

v_movement = np.vectorize(movement)
portfolio_20["Movement"] = v_movement(portfolio_20['Joint_movt'],
                                     portfolio_20['Previous_movt'])
portfolio_80["Movement"] = v_movement(portfolio_80['Joint_movt'],
                                     portfolio_80['Previous_movt'])

matrix_20 = portfolio_20['Movement'].value_counts()
matrix_80 = portfolio_80['Movement'].value_counts()

def probablity(self, total):
    result = self / total
    return result

def transition_table(matrix):
    result = pd.DataFrame(columns=["Moving Together", "Moving Apart"])
    result.at['Moving Together', 'Moving Together'] = probablity(matrix['mt_mt'], matrix.sum())
    result.at['Moving Together', 'Moving Apart'] = probablity(matrix['mt_ma'], matrix.sum())
    result.at['Moving Apart', 'Moving Together'] = probablity(matrix['ma_mt'], matrix.sum())
    result.at['Moving Apart', 'Moving Apart'] = probablity(matrix['ma_ma'], matrix.sum())
    return result
```

```
transition_table(matrix_80)
```

```
transition_table(matrix_20)
```

- How similar is the transition matrix from the first group to the second group?

In both groups, the transition matrix show us that the total probability of the transition table at each row is not equal to 1, thus making the transition matrix less helpful.

-  Is the process Markovian?

**The process is not Markovian:**

Because the sum of each row of the Moving Together and Moving Apart is not equal to 1, thus making it no Markovian process.
